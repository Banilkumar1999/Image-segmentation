# Importing the Required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
```

# Loading Data

```python
#Loading an image 
img=cv2.imread("cat.jpg");
```

```python
#Printing shape
img.shape
```

```python
#Asigning varieabel to shape
x,y,z = img.shape
```

```python
#printing data
plt.imshow(img)
plt.show()
```

# Converting BGR format to RGB fromat Image

```python
#Converting a image Color From BGR format to RGB format image
data =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#Printing RGB format image
plt.imshow(data)
plt.show()
```

# Re-shapeing image

```python
#Re-shaping image into 2D image
data=data.reshape((-1,3))
```

```python
#Printing 2D image shape
data.shape
```

# segmentation

```python
#Fiting a cluster model from KMeans
dominant_color = 2
model = KMeans(dominant_color)
model.fit(data)
```

```python
#applying a colour to model
colors = model.cluster_centers_
```

```python
colors = np.array(colors,dtype='uint8')
```

```python
#cheking array colors
colors
```

```python
#applying segmentation to 2d image
result= np.zeros((x*y,z),dtype='uint8')
```

```python
#setting color to center of the pixel level
for i in range(result.shape[0]):
    result[i]=colors[model.labels_[i]]
```

```python
#giving a Re-shape to Image
result = np.reshape(result,(x,y,z))
```

# Printing Final result.

```python
#Printing the final Segmentation image
plt.imshow(result)
plt.show()
```
